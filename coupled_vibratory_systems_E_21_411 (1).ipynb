QUESTION 1


To find the driving frequencies that cause violent vibrations (resonance), we need to determine the natural frequencies of the system. This is done by solving the eigenvalue problem for the undamped system:
$$ \det(K - \omega^2 M) = 0 $$

First, we will import the necessary numerical libraries and define the given system parameters.

```python
import numpy as np
from scipy.linalg import eigvals

# System Parameters
m = 1.6        # Mass (kg)
I = 0.14       # Mass moment of inertia (kg*m^2)
k = 1.9        # Spring stiffness (N/m)
l = 1.3        # Length (m)

# Calculate derived lengths
epsilon = 0.01 * l
l1 = l - epsilon
l2 = l + epsilon

print(f"Calculated l1: {l1:.4f} m")
print(f"Calculated l2: {l2:.4f} m")
```


Based on the equations of motion $M\ddot{Y} + C\dot{Y} + KY = F(t)$, we construct the mass matrix $M$ and the stiffness matrix $K$.
Since we are looking for the natural frequencies to determine resonance, we assume damping is negligible ($C \approx 0$).

$$ M = \begin{bmatrix} m & 0 \\ 0 & I \end{bmatrix} $$
$$ K = \begin{bmatrix} 2k & -k(l_2 - l_1) \\ -k(l_2 - l_1) & k(l_1^2 + l_2^2) \end{bmatrix} $$

```python
# Construct Mass Matrix (M)
M = np.array([
    [m, 0],
    [0, I]
])

# Construct Stiffness Matrix (K)
K = np.array([
    [2*k, -k*(l2 - l1)],
    [-k*(l2 - l1), k*(l1**2 + l2**2)]
])

print("Mass Matrix (M):")
print(M)
print("\nStiffness Matrix (K):")
print(K)
```


We solve the generalized eigenvalue problem $K\mathbf{v} = \lambda M\mathbf{v}$, where the eigenvalues $\lambda$ represent the square of the natural frequencies ($\lambda = \omega^2$).

We will calculate the eigenvalues, extract the natural frequencies, and identify the largest driving frequency that will cause violent vibrations.

```python
# Solve the generalized eigenvalue problem det(K - lambda*M) = 0
# Scipy's eigvals(A, B) solves det(A - lambda*B) = 0
eigenvalues = eigvals(K, M)

# The eigenvalues are purely real for symmetric positive definite M and K
# We take the real part to drop the +0j imaginary artifact from the solver
omega_squared = np.real(eigenvalues)

# Calculate natural frequencies (omega = sqrt(lambda))
omega = np.sqrt(omega_squared)

# Find the maximum frequency
max_omega = np.max(omega)

print(f"Calculated Natural Frequencies (omega): {omega}")
print(f"\nThe largest driving frequency for violent vibrations is: {max_omega:.4f} rad/s")
```

To find the driving frequencies that cause violent vibrations (resonance) in this 2-degree-of-freedom translational system, we need to determine its natural frequencies.

Violent vibrations occur when the driving frequency $\omega$ matches a natural frequency $\omega_n$. Since damping is described as "very small," we can assume $C \approx 0$ and solve the eigenvalue problem for the undamped system:
$$ \det(K - \omega^2 M) = 0 $$

First, let's import the necessary libraries and define the given parameters.

```python
import numpy as np
from scipy.linalg import eigvals

# System Parameters
m1 = 1.7       # Mass 1 (kg)
m2 = 1.4       # Mass 2 (kg)
k1 = 1.9       # Spring stiffness 1 (N/m)
k2 = 1.3       # Spring stiffness 2 (N/m)
k3 = 1.3       # Spring stiffness 3 (N/m)

print(f"Masses: m1 = {m1} kg, m2 = {m2} kg")
print(f"Stiffnesses: k1 = {k1} N/m, k2 = {k2} N/m, k3 = {k3} N/m")
```


Based on the provided equations, we construct the mass matrix $M$ and the stiffness matrix $K$.

*Note: While the generic formula in the image shows $M = \begin{bmatrix} m & 0 \\ 0 & m \end{bmatrix}$, specific values for $m_1$ and $m_2$ are given at the bottom. We will construct a general diagonal mass matrix using these specific values.*

$$ M = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix} $$
$$ K = \begin{bmatrix} k_1 + k_2 & -k_2 \\ -k_2 & k_2 + k_3 \end{bmatrix} $$

```python
# Construct Mass Matrix (M)
M = np.array([
    [m1, 0],
    [0, m2]
])

# Construct Stiffness Matrix (K)
K = np.array([
    [k1 + k2, -k2],
    [-k2, k2 + k3]
])

print("Mass Matrix (M):")
print(M)
print("\nStiffness Matrix (K):")
print(K)
```


We solve the generalized eigenvalue problem $K\mathbf{v} = \lambda M\mathbf{v}$, where the eigenvalues $\lambda$ represent the square of the natural frequencies ($\lambda = \omega^2$).

We will compute the eigenvalues, convert them to natural frequencies in rad/s, and extract the largest value to answer the question.

```python
# Solve the generalized eigenvalue problem det(K - lambda*M) = 0
eigenvalues = eigvals(K, M)

# The eigenvalues are purely real for symmetric positive definite M and K matrices
# We take the real part to drop the +0j imaginary artifact from the SciPy solver
omega_squared = np.real(eigenvalues)

# Calculate natural frequencies (omega = sqrt(lambda))
omega = np.sqrt(omega_squared)

# Find the maximum frequency
max_omega = np.max(omega)

print(f"Calculated Natural Frequencies (omega): {omega}")
print(f"\nThe largest driving frequency for violent vibrations is: {max_omega:.4f} rad/s")
```
